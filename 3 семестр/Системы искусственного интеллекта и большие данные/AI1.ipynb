4.3.1 –ó–∞–¥–∞–Ω–∏–µ. –ó–∞–¥–∞–π—Ç–µ  –º–∞—Å—Å–∏–≤ —Å–ª—É—á–∞–π–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –∏–∑ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞(0; 1).  
–†–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ —Å—Ä–µ–¥–Ω–∏–µ –∏ –º–µ–¥–∏–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –¥–ª—è –º–∞—Å—Å–∏–≤–∞, —Å—Ä–∞–≤–Ω–∏—Ç–µ 
—Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã, –∫–∞–∫–∏–µ –≤—ã–≤–æ–¥—ã –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –æ –∑–Ω–∞—á–µ–Ω–∏—è—Ö?  
–ü–æ—Å—Ç—Ä–æ–π—Ç–µ —Ç–æ—á–µ—á–Ω—É—é –¥–∏–∞–≥—Ä–∞–º–º—É —Ä–∞—Å—Å–µ–Ω–∏—è –ø–æ–ª—É—á–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞.

```python
import random
from statistics import mean, median
import matplotlib.pyplot as plt

array_size = 10
random_arr = [random.random() for _ in range(array_size)]

avg_value = mean(random_arr)
median_value = median(random_arr)

print(f"–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ: {avg_value:.4f}")
print(f"–ú–µ–¥–∏–∞–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ: {median_value:.4f}")

plt.scatter(range(len(random_arr)), random_arr, alpha=1, edgecolors='b', s=50)
plt.axhline(avg_value, color='r', linestyle='--', label=f"–°—Ä–µ–¥–Ω–µ–µ = {avg_value:.4f}")
plt.axhline(median_value, color='g', linestyle='-.', label=f"–ú–µ–¥–∏–∞–Ω–∞ = {median_value:.4f}")

plt.title("–î–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞—Å—Å–µ—è–Ω–∏—è —Å–ª—É—á–∞–π–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π")
plt.xlabel("–ò–Ω–¥–µ–∫—Å —ç–ª–µ–º–µ–Ω—Ç–∞")
plt.ylabel("–ó–Ω–∞—á–µ–Ω–∏–µ")
plt.legend()
plt.show()
```

4.3.2 –ó–∞–¥–∞–Ω–∏–µ. 
–°–æ–∑–¥–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ 10 –∑–Ω–∞—á–µ–Ω–∏–π 
—Ñ—É–Ω–∫—Ü–∏–∏ (ùë•, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç 1 –¥–æ 10). –í—ã–¥–µ–ª–∏—Ç—å —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π 
–ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ ‚Äì 
–ª–∏–Ω–µ–π–Ω—ã–π –∏ –¥–ª—è —Å—Ä–µ–∑–∞ ‚Äì —Ç–æ—á–µ—á–Ω—ã–π 

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 10 + 1)
y = (np.sqrt(1 + np.e**(np.sqrt(x)) + np.cos(x**2)) / (np.abs(1 - (np.sin(x))**3))) + np.log(2 * x)

from numpy import trapz
area = trapz(y, x)

# –ª–∏–Ω–µ–π–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è (x, y)
plt.figure(figsize=(10, 5))
plt.plot(x, y, c="r", label="–ò—Å—Ö–æ–¥–Ω—ã–π –º–∞—Å—Å–∏–≤")
plt.grid()
plt.title("–õ–∏–Ω–µ–π–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

x_slice = x[:5]
y_slice = y[:5]

# —Å—Ç—Ä–æ–∏–º —Ç–æ—á–µ—á–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è —Å—Ä–µ–∑–∞ –º–∞—Å—Å–∏–≤–∞
plt.figure(figsize=(10, 5))
plt.scatter(x_slice, y_slice, c="b", label="–°—Ä–µ–∑ –º–∞—Å—Å–∏–≤–∞", alpha=0.7)  # –¥–æ–±–∞–≤–ª—è–µ–º —Ç–æ—á–∫–∏ –¥–ª—è —Å—Ä–µ–∑–∞
plt.grid()
plt.title("–¢–æ—á–µ—á–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ –¥–ª—è —Å—Ä–µ–∑–∞ –º–∞—Å—Å–∏–≤–∞")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

print(f"–ü–ª–æ—â–∞–¥—å –ø–æ–¥ –≥—Ä–∞—Ñ–∏–∫–æ–º (–æ—Å–Ω–æ–≤–Ω–æ–π –º–∞—Å—Å–∏–≤): {area:.4f}")
```

4.3.3 –ó–∞–¥–∞–Ω–∏–µ. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫ –Ω–∞ 
–∏–Ω—Ç–µ—Ä–≤–∞–ª–µ (0,10) —Å —à–∞–≥–æ–º 1 —Å –∑–∞–ª–∏–≤–∫–æ–π –ø–ª–æ—â–∞–¥–∏ –∏ –Ω–∞–π—Ç–∏ —ç—Ç—É –ø–ª–æ—â–∞–¥—å 
–ø–æ–¥ –Ω–∏–º. –î–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –ø–ª–æ—â–∞–¥–∏ –≤–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏–∏ trapz(), 
–≤—ã—á–∏—Å–ª—è—é—â–µ–π –∏–Ω—Ç–µ–≥—Ä–∞–ª –ø–æ –ø—Ä–∞–≤–∏–ª—É —Ç—Ä–∞–ø–µ—Ü–∏–∏.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
from numpy import trapz

x = np.arange(0, 10 + 1)
y = np.abs(np.cos(x * np.e**(np.cos(x) + np.log(x + 1))))

area_trapz = trapz(y, x)
area_simps = simps(y, x)

plt.figure(figsize=(10, 6))
plt.fill_between(x, y, alpha=0.3, color='blue', label="–ü–ª–æ—â–∞–¥—å –ø–æ–¥ –≥—Ä–∞—Ñ–∏–∫–æ–º")
plt.title("–ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ —Å –∑–∞–ª–∏–≤–∫–æ–π –ø–ª–æ—â–∞–¥–∏")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()

print(f"–ü–ª–æ—â–∞–¥—å –ø–æ–¥ –≥—Ä–∞—Ñ–∏–∫–æ–º (–º–µ—Ç–æ–¥ —Ç—Ä–∞–ø–µ—Ü–∏–π): {area_trapz:.4f}")
print(f"–ü–ª–æ—â–∞–¥—å –ø–æ–¥ –≥—Ä–∞—Ñ–∏–∫–æ–º (–º–µ—Ç–æ–¥ –°–∏–º–ø—Å–æ–Ω–∞): {area_simps:.4f}")
```

4.3.5 –ó–∞–¥–∞–Ω–∏–µ. –°–æ–∑–¥–∞–π—Ç–µ –ø—Ä–æ—Å—Ç–µ–π—à–∏–π –∫–∞–ª—å–∫—É–ª—è—Ç–æ—Ä, –≤–∫–ª—é—á–∞—é—â–∏–π –æ—Å–Ω–æ–≤–Ω—ã–µ –¥–µ–π—Å—Ç–≤–∏—è 
–¥–ª—è –¥–≤—É—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö ‚Ä≤+‚Ä≤,‚Ä≤‚àí‚Ä≤,‚Ä≤‚àô‚Ä≤,‚Ä≤√∑‚Ä≤, –∞ —Ç–∞–∫–∂–µ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ 
—Å–ª–µ–¥—É—é—â–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π: ùëí^(ùë•+ùë¶),sin(ùë•+ùë¶),cos(ùë•+ùë¶),ùë•^ùë¶.

```python
import math

def calculator():
    print("–í–≤–µ–¥–∏—Ç–µ –¥–≤–∞ —á–∏—Å–ª–∞ (x –∏ y):")
    x = float(input("x = "))
    y = float(input("y = "))
    
    print("\n–í—ã–±–µ—Ä–∏—Ç–µ –æ–ø–µ—Ä–∞—Ü–∏—é:")
    print("1. –°–ª–æ–∂–µ–Ω–∏–µ (+)")
    print("2. –í—ã—á–∏—Ç–∞–Ω–∏–µ (-)")
    print("3. –£–º–Ω–æ–∂–µ–Ω–∏–µ (*)")
    print("4. –î–µ–ª–µ–Ω–∏–µ (/)")
    print("5. –í–æ–∑–≤–µ–¥–µ–Ω–∏–µ –≤ —Å—Ç–µ–ø–µ–Ω—å (x^y)")
    print("6. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ e^(x+y)")
    print("7. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ sin(x+y)")
    print("8. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ cos(x+y)")
    
    choice = int(input("\n–í–≤–µ–¥–∏—Ç–µ –Ω–æ–º–µ—Ä –æ–ø–µ—Ä–∞—Ü–∏–∏ (1-8): "))
    
    if choice == 1:
        result = x + y
        print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: {x} + {y} = {result}")
    elif choice == 2:
        result = x - y
        print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: {x} - {y} = {result}")
    elif choice == 3:
        result = x * y
        print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: {x} * {y} = {result}")
    elif choice == 4:
        if y != 0:
            result = x / y
            print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: {x} / {y} = {result}")
        else:
            print("–û—à–∏–±–∫–∞: –î–µ–ª–µ–Ω–∏–µ –Ω–∞ –Ω–æ–ª—å!")
    elif choice == 5:
        result = x ** y
        print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: {x}^{y} = {result}")
    elif choice == 6:
        result = math.exp(x + y)
        print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: e^({x} + {y}) = {result}")
    elif choice == 7:
        result = math.sin(x + y)
        print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: sin({x} + {y}) = {result}")
    elif choice == 8:
        result = math.cos(x + y)
        print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç: cos({x} + {y}) = {result}")
    else:
        print("–û—à–∏–±–∫–∞: –ù–µ–≤–µ—Ä–Ω—ã–π –Ω–æ–º–µ—Ä –æ–ø–µ—Ä–∞—Ü–∏–∏!")

calculator()
```
