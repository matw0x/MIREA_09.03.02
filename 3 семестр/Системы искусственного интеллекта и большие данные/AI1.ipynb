4.3.1 Задание. Задайте  массив случайных значений из интервала(0; 1).  
Рассчитайте средние и медианные значения для массива, сравните 
результаты, какие выводы можно сделать о значениях?  
Постройте точечную диаграмму рассения полученного ряда.

```python
import random
from statistics import mean, median
import matplotlib.pyplot as plt

array_size = 10
random_arr = [random.random() for _ in range(array_size)]

avg_value = mean(random_arr)
median_value = median(random_arr)

print(f"Среднее значение: {avg_value:.4f}")
print(f"Медианное значение: {median_value:.4f}")

plt.scatter(range(len(random_arr)), random_arr, alpha=1, edgecolors='b', s=50)
plt.axhline(avg_value, color='r', linestyle='--', label=f"Среднее = {avg_value:.4f}")
plt.axhline(median_value, color='g', linestyle='-.', label=f"Медиана = {median_value:.4f}")

plt.title("Диаграмма рассеяния случайных значений")
plt.xlabel("Индекс элемента")
plt.ylabel("Значение")
plt.legend()
plt.show()
```

4.3.2 Задание. 
Создать массив из 10 значений 
функции (𝑥, например, изменяется от 1 до 10). Выделить срез первой 
половины массива и построить графики для основного массива – 
линейный и для среза – точечный 

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 10 + 1)
y = (np.sqrt(1 + np.e**(np.sqrt(x)) + np.cos(x**2)) / (np.abs(1 - (np.sin(x))**3))) + np.log(2 * x)

from numpy import trapz
area = trapz(y, x)

# линейный график для (x, y)
plt.figure(figsize=(10, 5))
plt.plot(x, y, c="r", label="Исходный массив")
plt.grid()
plt.title("Линейный график функции")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

x_slice = x[:5]
y_slice = y[:5]

# строим точечный график для среза массива
plt.figure(figsize=(10, 5))
plt.scatter(x_slice, y_slice, c="b", label="Срез массива", alpha=0.7)  # добавляем точки для среза
plt.grid()
plt.title("Точечный график для среза массива")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

print(f"Площадь под графиком (основной массив): {area:.4f}")
```

4.3.3 Задание. Построить график на 
интервале (0,10) с шагом 1 с заливкой площади и найти эту площадь 
под ним. Для вычисления площади воспользуйте функции trapz(), 
вычисляющей интеграл по правилу трапеции.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
from numpy import trapz

x = np.arange(0, 10 + 1)
y = np.abs(np.cos(x * np.e**(np.cos(x) + np.log(x + 1))))

area_trapz = trapz(y, x)
area_simps = simps(y, x)

plt.figure(figsize=(10, 6))
plt.fill_between(x, y, alpha=0.3, color='blue', label="Площадь под графиком")
plt.title("График функции с заливкой площади")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()

print(f"Площадь под графиком (метод трапеций): {area_trapz:.4f}")
print(f"Площадь под графиком (метод Симпсона): {area_simps:.4f}")
```

4.3.5 Задание. Создайте простейший калькулятор, включающий основные действия 
для двух переменных ′+′,′−′,′∙′,′÷′, а также вычисление 
следующих функций: 𝑒^(𝑥+𝑦),sin(𝑥+𝑦),cos(𝑥+𝑦),𝑥^𝑦.

```python
import math

def calculator():
    print("Введите два числа (x и y):")
    x = float(input("x = "))
    y = float(input("y = "))
    
    print("\nВыберите операцию:")
    print("1. Сложение (+)")
    print("2. Вычитание (-)")
    print("3. Умножение (*)")
    print("4. Деление (/)")
    print("5. Возведение в степень (x^y)")
    print("6. Вычисление e^(x+y)")
    print("7. Вычисление sin(x+y)")
    print("8. Вычисление cos(x+y)")
    
    choice = int(input("\nВведите номер операции (1-8): "))
    
    if choice == 1:
        result = x + y
        print(f"Результат: {x} + {y} = {result}")
    elif choice == 2:
        result = x - y
        print(f"Результат: {x} - {y} = {result}")
    elif choice == 3:
        result = x * y
        print(f"Результат: {x} * {y} = {result}")
    elif choice == 4:
        if y != 0:
            result = x / y
            print(f"Результат: {x} / {y} = {result}")
        else:
            print("Ошибка: Деление на ноль!")
    elif choice == 5:
        result = x ** y
        print(f"Результат: {x}^{y} = {result}")
    elif choice == 6:
        result = math.exp(x + y)
        print(f"Результат: e^({x} + {y}) = {result}")
    elif choice == 7:
        result = math.sin(x + y)
        print(f"Результат: sin({x} + {y}) = {result}")
    elif choice == 8:
        result = math.cos(x + y)
        print(f"Результат: cos({x} + {y}) = {result}")
    else:
        print("Ошибка: Неверный номер операции!")

calculator()
```
